```python
# Mostra parâmetros da ANOVA
an_ataque

# A hipótese nula na ANOVA é que todas as médias dos grupos são iguais.

# O valor F na Análise de Variância (ANOVA) é uma estatística de teste que é usada
# para comparar as variâncias entre os grupos.
### A estatística F é uma razão de duas medidas de variação nos dados.
### >>> Se a hipótese nula for verdadeira, ambas são estimativas da mesma coisa e a proporção será em torno de 1. <<<<<
# O numerador é calculado medindo a variação das médias e se as médias verdadeiras
# dos grupos são idênticas, isso é uma função da variação geral dos dados.
# Mas se a hipótese nula for falsa e os meios não forem todos iguais, essa medida
# de variação será maior.
# O denominador é uma média das variações da amostra para cada grupo, que é uma
# estimativa da variação geral da população (assumindo que todos os grupos tenham
# variações iguais).
# Portanto, quando o nulo de todos os meios iguais for verdadeiro, as 2 medidas
# serão semelhantes e a proporção será próxima de 1.
# Se H_0 for falso, o numerador será grande em relação ao denominador e a razão
# serão maiores que 1.

# PR(>F) na Análise de Variância (ANOVA) é o valor-p associado ao valor F calculado no teste.
# O valor-p é uma medida de significância estatística e é usado para decidir se a hipótese nula
# pode ser rejeitada.
### Se o valor-p for menor que o nível de significância escolhido (geralmente 0,05),
### então você rejeita a hipótese nula e conclui que pelo menos uma das médias dos grupos é
### significativamente diferente.
# Portanto, PR(>F) é a probabilidade de obter um valor F pelo menos tão extremo quanto o
# que foi calculado a partir dos dados, assumindo que a hipótese nula seja verdadeira.
# Se PR(>F) for pequeno (geralmente menor que 0,05), isso indica que é improvável que
# as diferenças observadas nos dados tenham ocorrido por acaso, e a hipótese nula é rejeitada.
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
# from sklearn.model_selection import train_test_split


# Conjunto de dados de teste e traino
X, y = make_classification(n_samples=100, n_features=6, n_informative=3, n_redundant=0, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# Crie uma instância do MLPClassifier
clf = MLPClassifier(hidden_layer_sizes=(1, 5), max_iter=1000, alpha=1e-4,
                    solver='sgd', verbose=10, random_state=42,
                    learning_rate_init=.1)

# quanto ao número de nos, recomenda-se usar de 0,5 a 2 vezes o número de features.

# - hidden_layer_sizes: Este parâmetro permite que você defina o número de camadas
# ocultas, bem como o número de neurônios em cada camada. Ele aceita uma tupla,
# onde o comprimento da tupla representa o número de camadas ocultas e cada número
# na tupla representa o número de neurônios naquela camada.

# - alpha: Este é o parâmetro de regularização L2. Ele ajuda a evitar o sobreajuste
# adicionando uma penalidade aos pesos durante o treinamento. Um valor maior de
# alpha resulta em pesos menores, criando um modelo mais simples.

# - verbose: Se definido como True, o algoritmo imprimirá mensagens de progresso
# (como a perda de treinamento) enquanto treina. Isso pode ser útil para depuração
#  ou para entender o progresso do treinamento.

# - random_state: Este parâmetro controla a aleatoriedade do inicializador de pesos.
# Se você definir um número inteiro para random_state, então você pode garantir
# que seu código produzirá a mesma saída cada vez que for executado com essa mesma
# entrada.

# - learning_rate_init: Este é o tamanho inicial do passo para o otimizador de peso.
# Se o learning_rate for definido como ‘constant’, então o learning_rate_init será
# usado durante todo o treinamento para atualizar os pesos. Se o learning_rate for
# definido como ‘adaptive’ ou ‘invscaling’, então o learning_rate_init será
# ajustado ao longo do treinamento1.

# Treine o modelo
clf.fit(X_train, y_train)


# Avalie o modelo
print("Acurácia do treinamento: %f" % clf.score(X_train, y_train))
print("Acurácia do teste: %f" % clf.score(X_test, y_test))
```
