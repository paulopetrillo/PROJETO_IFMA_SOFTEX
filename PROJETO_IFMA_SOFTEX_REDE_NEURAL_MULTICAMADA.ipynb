# Primeiro Treinamento

```python
# Carrega as bibliotecas
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
# X é o dataframe sem a coluna de saída
X = dataframe_escolhido.drop("classe_resultado", axis=1)
print(X.head())


 # y é a coluna de saída
y = dataframe_escolhido["classe_resultado"]
print("----")
print(y.head())
```

```python
X
```

```python
y.value_counts()
```

```python
# Definir as variáveis ​​preditoras e de resposta
features = X
target = y

 # # Conjunto de dados de teste e traino
X_train, X_test, y_train, y_test = train_test_split(features,
                                                    target,
                                                    test_size=0.25,
                                                    random_state=42)
```

```python
# Cria uma instância do MLPClassifier
mlp = MLPClassifier(max_iter=500
                  , activation='tanh'
                  , alpha=0.0001
                  , hidden_layer_sizes=(50,)
                  , learning_rate='adaptive'
                  , solver='sgd'
                  , verbose=0
                  , random_state=42)
```

```python
# Treinamento do modelo
#   mlp =    mlp.fit(X_train, y_train)
mlp = mlp.fit(X_train, y_train)

# # Imprime Acurácia do modelo
print("Acurácia do treinamento: %f" % mlp.score(X_train, y_train))
print("Acurácia do teste: %f" % mlp.score(X_test, y_test))
```

```python
import pickle

# Salva o modelo
with open('mlp_model_etapa_1.pkl', 'wb') as f:
    pickle.dump(mlp, f)
```

```python
#imprimir o relatório de classificação para MLP (Machine Learn Perception)
print(classification_report(y_test, mlp.predict(X_test)))
```

```python

# Os métodos de balanceamento de carga em Python são técnicas que visam corrigir o desbalanceamento de classes em um conjunto de dados.
# O desbalanceamento de classes ocorre quando uma classe é representada de forma desproporcional às demais classes.
# Isso pode prejudicar o desempenho de um modelo de aprendizado de máquina, pois o modelo pode aprender a identificar apenas a classe mais representada.

# Os métodos de balanceamento de carga podem ser divididos em dois grupos principais:
# Métodos de reamostragem: Esses métodos alteram o conjunto de dados para que as classes sejam mais balanceadas.
# Métodos de ponderação: Esses métodos atribuem pesos diferentes às amostras de cada classe, de modo que as classes menos representadas tenham mais importância.
# Métodos de reamostragem

# Os métodos de reamostragem são os mais comuns para balancear conjuntos de dados.
# Esses métodos podem ser divididos em dois subgrupos:
# **Reamostragem upsampling: Esse método aumenta o número de amostras da classe menos representada. Isso pode ser feito copiando amostras existentes ou gerando novas amostras.
# Reamostragem downsampling: Esse método reduz o número de amostras da classe mais representada. Isso pode ser feito removendo amostras ou subamostrando amostras.

# Métodos de ponderação
# Os métodos de ponderação atribuem pesos diferentes às amostras de cada classe.
# Isso faz com que o modelo atribua mais importância às amostras das classes menos representadas.

# Exemplos de métodos de balanceamento de carga em Python
# A biblioteca scikit-learn fornece uma variedade de métodos de balanceamento de carga. Alguns exemplos incluem:

# Undersample: Esse método usa a técnica de reamostragem downsampling.
# Oversample: Esse método usa a técnica de reamostragem upsampling.
# SMOTE: Esse método é uma técnica de reamostragem upsampling que gera novas amostras a partir de amostras existentes.
# Tomek links: Esse método é uma técnica de reamostragem downsampling que identifica e remove amostras da classe mais representada que são facilmente classificadas como pertencentes à classe menos representada.
# Cost-sensitive learning: Esse método é uma técnica de ponderação que atribui pesos diferentes às amostras de cada classe.


# Como escolher um método de balanceamento de carga

# A escolha do método de balanceamento de carga depende de vários fatores, incluindo:
# O grau de desbalanceamento do conjunto de dados: Se o desbalanceamento for muito grande, pode ser necessário usar um método mais agressivo.
# O tipo de modelo de aprendizado de máquina: Alguns modelos são mais sensíveis ao desbalanceamento do que outros.
# Os recursos disponíveis: Alguns métodos de balanceamento de carga são mais computacionalmente caros do que outros.
# Em geral, é recomendável experimentar diferentes métodos de balanceamento de carga para encontrar o que funciona melhor para o seu conjunto de dados e modelo específico.
```
