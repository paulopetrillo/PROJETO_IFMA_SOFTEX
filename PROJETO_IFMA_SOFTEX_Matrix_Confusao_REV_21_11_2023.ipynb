```python
dataframe_escolhido.drop(['mandante', 'golsMand', 'golsVist', 'vistante'], axis=1, inplace=True)
dataframe_escolhido
```

```python
dataframe_escolhido = dataframe_escolhido[['poder_ataque_time_mandante', 'poder_ataque_time_visitante', 'poder_defesa_time_mandante',	'poder_defesa_time_visitante',	'classe_resultado' ]]
dataframe_escolhido
```

```python
# Carrega as bibliotecas
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
# # Estabele o conjunto de dados de treinamento.
# X = dataframe_escolhido.drop('classe_resultado', axis=1)
# y = dataframe_escolhido['classe_resultado']  # y é a coluna de saída

  # X é o dataframe sem a coluna de saída
X = dataframe_escolhido.drop("classe_resultado", axis=1)
print(X.head())
 # y é a coluna de saída
y = dataframe_escolhido["classe_resultado"]
print("----")
print(y.head())
```

```python
'''
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=0)
X_resampled, y_resampled = rus.fit_resample(X, y)
#y_resampled.value_counts()
'''
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
X_resampled, y_resampled = ros.fit_resample(X, y)
#y_resampled.value_counts()
```

```python
# Divide os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X_resampled,
                                                    y_resampled,
                                                #    random_state=42,
                                                    test_size=0.2
                                                    )
```

```python
# Visualizando/conferinfo
#type(X_train)
#print(X_train)
#print(y_train)
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

```python
from sklearn.model_selection import GridSearchCV
mlp = MLPClassifier()
parameter_space = {
    'hidden_layer_sizes': [(50,), (5,), (3,)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant','adaptive'],
}
clf = GridSearchCV(mlp, parameter_space, n_jobs=-1, cv=3)
clf.fit(X_train, y_train)
print('Best parameters found:\n', clf.best_params_)
```

```python
# Cria uma instância do MLPClassifier
mlp = MLPClassifier(max_iter=500,
                    hidden_layer_sizes=(50,) ,
                    activation='tanh',
                    alpha=0.05,
                    learning_rate='constant',
                    solver='sgd',
                    verbose=10,
                    random_state=42)

# Treinamento do modelo
mlp = mlp.fit(X_train, y_train)


# Imprime Acurácia do modelo
print("Acurácia do treinamento: %f" % mlp.score(X_train, y_train))
print("Acurácia do teste: %f" % mlp.score(X_test, y_test))
```
