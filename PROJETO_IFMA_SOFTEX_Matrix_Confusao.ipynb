```python
# Copia dataframe.
dataframe_escolhido = DataFrame_Geral.copy()
dataframe_escolhido
```

```python
dataframe_escolhido.drop(['mandante', 'golsMand', 'golsVist', 'vistante'], axis=1, inplace=True)
dataframe_escolhido
```

```python
dataframe_escolhido = dataframe_escolhido[['poder_ataque_time_mandante', 'poder_ataque_time_visitante', 'poder_defesa_time_mandante',	'poder_defesa_time_visitante',	'classe_resultado' ]]
dataframe_escolhido
```

```python
# Carrega as bibliotecas
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
# # Estabele o conjunto de dados de treinamento.
# X = dataframe_escolhido.drop('classe_resultado', axis=1)
# y = dataframe_escolhido['classe_resultado']  # y é a coluna de saída

  # X é o dataframe sem a coluna de saída
X = dataframe_escolhido.drop("classe_resultado", axis=1)

 # y é a coluna de saída
y = dataframe_escolhido["classe_resultado"]
```

```python
# Divide os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                #    random_state=42,
                                                    test_size=0.2
                                                    )
```

```python
# Visualizando/conferinfo
#type(X_train)
#print(X_train)
#print(y_train)
#print(len(X_train))
#print(len(y_train))
```

```python
# Cria uma instância do MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(1, 5),
                    max_iter=1000,
                    alpha=1e-4,
                    solver='sgd',
                    verbose=10,
                    random_state=42,
                    learning_rate_init=.1)

# Treinamento do modelo
mlp.fit(X_train, y_train)


# Imprime Acurácia do modelo
print("Acurácia do treinamento: %f" % mlp.score(X_train, y_train))
print("Acurácia do teste: %f" % mlp.score(X_test, y_test))
```

```python
# Previsão baseada nos parametrosdo modelo.
predictions = mlp.predict(X_test)

print(predictions)
```
